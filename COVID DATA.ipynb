# Group by :

 It is used to summarise the large amount of data with the average function.

```python
df["Country/Region"].nunique()
```

```python
x = df.groupby("Country/Region")["Confirmed"].sum().reset_index()  # reset_index - 
x
```

```python
# How to sort the value in pandas

x = x.sort_values("Confirmed", ascending=False)
x.head(10)
```

```python
# Find out last five country where confirmed counts is lowest
x = x.sort_values("Confirmed", ascending=False)
x.tail(5)
```

```python
# show the top 10 countries base on death

df["Country/Region"].nunique()
x = df.groupby("Country/Region")["Deaths"].sum().reset_index()
x = x.sort_values("Deaths", ascending=False)
x.head(10)
```

```python
# show the top 10 countries base on Active case

df["Country/Region"].nunique()
x = df.groupby("Country/Region")["Active"].sum().reset_index()
x = x.sort_values("Active", ascending=False)
x.head(10)
```

```python
# find out active, deaths & confirmed cases for all countries.

x = df.groupby("Country/Region")[["Active","Deaths","Confirmed"]].sum().reset_index()
x
```

```python
# How to add the new columns
# new columns add on top right column

df["Confirmed_case"] = df["Confirmed"]+df["New cases"]       # it add new column Confirmed_Case on top right column
df["Death_case"] = df["Deaths"]+df["New deaths"]             # it add new column Death_Case on top right column
df["Recovered_case"] = df["Recovered"]+df["New recovered"]   # it add new column Recovered_case on top right column
df.head()
```

```python
# How to delete the columns.
# delete Confirmed col.
# 1st Methord

df.drop(columns="Confirmed", inplace=True)
df
```

```python
# How to delete the columns.
# delete Deaths col.
# 2nd Methord

df.drop("Deaths",axis=1, inplace=True)
df.head()
```
